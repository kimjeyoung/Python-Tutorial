### 사칙연산

파이썬에서 덧셈, 뺄셈, 나눗셈, 곱셈은 다음과 같음
```python
a = 10
b = 3

# 덧셈
add = a + b
# 뺄셈
subtract = a - b
# 나눗셈
division = a / b
# 곱셈
multiply = a * b
```

우리는 많은 경우에 동일한 변수에 특정 값을 더하거나 뺴거나 곱하거나 나누는 경우가 있음. 예를 들어 a=2일때 a에 3을 더한 값을 만들어야 하는경우. 
그럴 때에는 파이썬을 이용할 때 두가지 방식으로 구현될 수 있음 
```python
# case 1
a = 2
a = a + 2

# case 2 
a += 2
```

어떤 코드가 더 좋다라는 정답은 없지만 대부분 case2로 코딩함. "a = a + 2" 의 의미는 "a += 2" 와 동일함. 마찬가지로 덧셈뿐만 아니라 뺄셈, 나눗셈, 곱셈도 가능 (e.g., "-=", "/=", "*=")


* 나눗셈에서 몫과 나머지를 구하는 방법

```python 
# 몫
quotient = a // b
# 나머지
remainder = a % b
```

### 출력

파이썬에서 특정 값들을 출력해서 확인하고 싶다면 "print()" 함수를 사용하면 됨.

```python

# example 01
a = 10
print(a)

# example 02
txt = "hello world!"
print(txt)

```


### 함수

파이썬에서 함수란 반복적으로 사용하는 특정기능을 효율적으로 관리하기 위해서 사용하는 기능중에 하나. 개발자가 함수를 작성할 때 함수안에 들어갈 input(**매개변수**), output을 설계해야함.

아래 예시에서 반복문 (for i in range(100)) 이 반복적으로 사용됨. 이때 개별적으로 num1, num2에 대한 값을 구하는 기능(덧셈)은 통합되어 관리될 수 있음. 


```python
# bad example
num1 = 0
for i in range(100):
    num1 += 1

num2 = 10
for i in range(100):
    num2 += 1
```

위의 코드를 함수화 하게 되면 아래와 같이 특정 "덧셈" 기능을 담당하는 함수로 만들 수 있음. 
함수화, 객체화를 클린코드 (기업의 입장에선 코드의 품질 관리 및 유지보수)를 위해서 매우 중요하기 때문에 꼭 알아야함.

```python

def add_function(x):
    for i in range(100):
        x += 1
    return x

num1 = 0
num2 = 10

# 함수를 호출하면 "add_function" 이 실행되고 return 오른쪽값이 출력되는 형태
num1 = add_function(num1)
num2 = add_function(num2)
```

### 반복문

반복되는 연산을 해야할 경우에 사용할 수 있음. 예를 들어, 초기값이 3인 x 에 100번 2를 더하는 기능을 만들어야 한다고 할때, 파이썬을 이용하면 다음과 같이 코딩할 수 있음. 

```python
x = 3
for i in range(100):
    x = x + 2
```
